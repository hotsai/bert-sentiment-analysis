# Fetch Annual Reports

```python
#%% 1. Import packages
import time
from tqdm import tqdm
import pandas as pd

# Required installation: https://pypi.org/project/selenium/
# Specify version: pip install selenium==3.141.0
# ActionChains: Simulate mouse and keyboard events, parameters: https://www.selenium.dev/selenium/docs/api/py/webdriver/selenium.webdriver.common.action_chains.html
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options

# Required installation: https://pypi.org/project/webdriver-manager/
from webdriver_manager.chrome import ChromeDriverManager

#%% 2. Settings

# Setting - Time sleep
# In seconds
t_sleep = 20

# Read Year and ID number
data = pd.read_excel("/Users/hohan/Desktop/company_list.xlsx")

#%% 3. Scrape

list_succ = []
list_fail = []

# for year in tqdm(year_list):
    # for id_number in tqdm(range(Stock_Code_Beg, Stock_Code_End)):
for i in tqdm(range(data.shape[0])):
    year = str(int(data['year'][i]) - 1911)  # Transfer 2016 to 105
    id_number = str(int(data['id'][i]))

    # year = 104
    # id_number = 1101

    try:
        driver = webdriver.Chrome(ChromeDriverManager().install())

        url = f"https://doc.twse.com.tw/server-java/t57sb01?step=1&colorchg=1&co_id={id_number}&year={year}&mtype=F&"
        driver.get(url)

        # Get or dispose of asset handling process (establish or revise)
        link = driver.find_elements_by_xpath('//a[contains(., "F04") and contains(@href, ".pdf")]')[0].get_attribute(
            'href')
        # Get the PDF name
        pdf_name = link.split('\"')[-2]
        # Simulate clicking on the PDF file
        driver.find_element_by_xpath(f"//a[contains(@href,'{link}')]").click()

        # Get all current browser windows
        windows = driver.window_handles
        # Switch to the latest window
        driver.switch_to.window(windows[-1])
        # Sleep for 3 seconds
        time.sleep(3)
        # Get the current URL of the webpage
        url = driver.current_url

        # Simulate clicking the PDF file again to generate the PDF on the Public Information Observation Station for subsequent download
        right_click = driver.find_element_by_link_text(pdf_name)
        # Left-click
        ActionChains(driver).double_click(right_click).perform()

        # Get all current browser windows
        windows = driver.window_handles
        # Switch to the latest window
        driver.switch_to.window(windows[-1])
        # Sleep for 3 seconds
        time.sleep(1)
        # Check the current URL
        url = driver.current_url
        driver.quit()

        chrome_options = Options()
        chrome_options.add_experimental_option(
            'prefs', {"download.default_directiory": './',
                      'profile.default_content_settings.popups': 0,
                      "download.prompt_for_download": False,
                      "download.directory_upgrade": True,
                      "plugins.always_open_pdf_externally": True})
        browser = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
        browser.get(url)
        time.sleep(t_sleep)
        # browser.quit()

        # print(year, id_number, "Successed!")
        list_succ.append(str(year) + str(id_number))

    except:
        # print(year, id_number, "Failed!")
        list_fail.append(str(year) + str(id_number))
        try:
            browser.quit()
        except:
            pass
```
